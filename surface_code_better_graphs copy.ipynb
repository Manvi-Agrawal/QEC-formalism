```python
import pandas as pd
```

```python
def get_time_elapsed(sc):
    sc["config"] += sc["init"]
    sc["cons_gen"] += sc["config"]
    sc["cons_sol"] += sc["cons_gen"]
    # sc.show()
    return sc
```

## Buggy

```python
sc_buggy = pd.read_csv("surface_code_buggy.csv")

sc_buggy
```

```python
sc_buggy_t = get_time_elapsed(sc_buggy)
sc_buggy_t
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

fig, ax = plt.subplots()
plt.rcParams['axes.titlepad'] = 10


# plt.rcParams.update({'font.size': 20})
ax.plot(sc_buggy_t["d"], sc_buggy_t["init"], marker="o", label="init")
ax.plot(sc_buggy_t["d"], sc_buggy_t["config"], marker="o", label="config")
ax.plot(sc_buggy_t["d"], sc_buggy_t["cons_gen"], marker="o", label="cons_gen")
ax.plot(sc_buggy_t["d"], sc_buggy_t["cons_sol"], marker="o", label="cons_sol")



ax.set_title("Buggy Decoder: Execution Time Visualization", fontsize=18)
ax.set_xlabel("Code distance (d)", fontsize=14)
ax.set_ylabel("Execution Time(in sec)", fontsize=14)

# plt.legend()
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width, pos.height * 0.85])
ax.legend(
    # loc='upper center', 
    # bbox_to_anchor=(0.5, 1.35),
    # ncol=4, 
    fontsize=12,
)

plt.show()

plt.rcParams.update(plt.rcParamsDefault)
```

## Execution Time Comparison

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

fig, ax = plt.subplots(nrows=2, ncols=2, sharex="col", figsize=(8,8))
plt.rcParams.update(plt.rcParamsDefault)
# plt.rcParams.update({'font.size': 16})
# plt.rcParams['axes.titlepad'] = 20

ax[0,0].plot(sc_buggy_t["d"], sc_buggy_t["init"], label="init", marker="o")
ax[0,0].plot(sc_correct_t["d"], sc_correct_t["init"], label="init", marker="o")
ax[0,0].set_title("Initialisation", fontsize=14)

ax[0,1].plot(sc_buggy_t["d"], sc_buggy_t["config"], label="config", marker="o")
ax[0,1].plot(sc_correct_t["d"], sc_correct_t["config"], label="config", marker="o")
ax[0,1].set_title("Config setup", fontsize=14)


ax[1,0].plot(sc_buggy_t["d"], sc_buggy_t["cons_gen"], label="cons_gen", marker="o")
ax[1,0].plot(sc_correct_t["d"], sc_correct_t["cons_gen"], label="cons_gen", marker="o")
ax[1,0].set_title("Constraint generation", fontsize=14)


ax[1,1].plot(sc_buggy_t["d"], sc_buggy_t["cons_sol"], label="buggy", marker="o")
ax[1,1].plot(sc_correct_t["d"], sc_correct_t["cons_sol"], label="correct", marker="o")
ax[1,1].set_title("Constraint solving", fontsize=14)



fig.suptitle("Visualize stages: Buggy vs Correct", fontsize=20)
fig.supxlabel("Code distance (d)", fontsize=16)
fig.supylabel("Execution Time(in sec)", fontsize=16)


lines, labels = ax[1, 1].get_legend_handles_labels()
# fig.legend(handles, labels, ncol=1)
plt.figlegend(lines, labels, loc="upper right")


# loc='upper center'
 
plt.show()

plt.rcParams.update(plt.rcParamsDefault)

```
