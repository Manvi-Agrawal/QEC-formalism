```python
import pandas as pd
```

```python
sc_buggy = pd.read_csv("surface_code_buggy.csv")

sc_buggy
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

fig, ax = plt.subplots()

plt.rcParams.update({'font.size': 20})

plt.plot(sc_buggy["d"], sc_buggy["all"], label="buggy")
plt.scatter(sc_buggy["d"], sc_buggy["all"], label="buggy")

plt.title("Finding Buggy Rotatated Surface Code")
plt.xlabel("Code distance (d)")
plt.ylabel("Execution Time(in sec)")

# plt.legend()
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

fig, ax = plt.subplots()

plt.rcParams.update({'font.size': 20})

plt.plot(sc_buggy["nq"], sc_buggy["all"], label="buggy")
plt.scatter(sc_buggy["nq"], sc_buggy["all"], label="buggy")

plt.title("Finding Buggy Rotatated Surface Code")
plt.xlabel("Number of physical qubits (nq)")
plt.ylabel("Execution Time(in sec)")

# plt.legend()
```
