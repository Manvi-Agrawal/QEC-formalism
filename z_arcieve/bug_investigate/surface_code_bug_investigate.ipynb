```python
def get_time_elapsed(sc):
    sc["config"] += sc["init"]
    sc["cons_gen"] += sc["config"]
    sc["cons_sol"] += sc["cons_gen"]
    # sc.show()
    return sc
```

## Buggy

```python
sc_buggy = pd.read_csv("surface_code_buggy.csv")

sc_buggy
```

```python
sc_typo_buggy = pd.read_csv("rsc_typo_buggy.csv")

sc_typo_buggy
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

fig = plt.figure()
plt.rcParams['axes.titlepad'] = 10


# plt.rcParams.update({'font.size': 20})
plt.plot(sc_buggy["d"], sc_buggy["all"], marker="o", label="data")
plt.plot(sc_typo_buggy["d"], sc_typo_buggy["all"], marker="o", label="typo")




plt.title("Verifying Rotatated Surface Code", fontsize=18)
plt.xlabel("Code distance (d)", fontsize=14)
plt.ylabel("Execution Time(in sec)", fontsize=14)

plt.legend(fontsize=12)

plt.show()
plt.rcParams.update(plt.rcParamsDefault)


```
