```python
import pandas as pd
```

```python
toric_code = pd.read_csv("toric_code.dat", delimiter=" ")
# toric_code = pd.read_csv("toric_code.csv")
toric_code
```

```python
import numpy as np
toric_code["d"] = np.int32(np.sqrt(toric_code["nq"]/2))
toric_code
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

plt.rcParams['axes.titlepad'] = 10


fig, ax = plt.subplots()

# plt.rcParams.update({'font.size': 20})
# plt.scatter(toric_code["nq"], toric_code["all"])
plt.plot(toric_code["nq"], toric_code["all"], marker="o")

plt.title("Verifying Toric code", fontsize=20)
plt.xlabel("Code distance (d)",  fontsize=16)
plt.ylabel("Execution Time(in sec)",  fontsize=16)

plt.show()

plt.rcParams.update(plt.rcParamsDefault)
```

```python
sc_buggy = pd.read_csv("surface_code_buggy.csv")

sc_buggy
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

plt.rcParams['axes.titlepad'] = 10

plt.plot(toric_code["nq"], toric_code["all"], marker="o", label="toric code")
plt.plot(sc_buggy["nq"], sc_buggy["all"], marker="o", label="surface code")


plt.title("Comparing Execution Times", fontsize=18)
plt.xlabel("Number of physical qubits (nq)", fontsize=14)

plt.ylabel("Execution Time(in sec)", fontsize=14)

plt.legend(fontsize=12, loc="upper left")

plt.show()
plt.rcParams.update(plt.rcParamsDefault)
```
