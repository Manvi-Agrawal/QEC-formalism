```python
import pandas as pd
```

```python
def get_time_elapsed(sc):
    sc["config"] += sc["init"]
    sc["cons_gen"] += sc["config"]
    sc["cons_sol"] += sc["cons_gen"]
    # sc.show()
    return sc
```

## Buggy

```python
sc_buggy = pd.read_csv("surface_code_buggy.csv")

sc_buggy
```

```python
sc_buggy_t = get_time_elapsed(sc_buggy)
sc_buggy_t
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')
plt.rcParams['axes.titlepad'] = 10


# fig, ax = plt.subplots()
fig, ax1 = plt.subplots(figsize=(4.8, 3))

color1 = 'tab:blue'

ax1.plot(sc_buggy_t["d"], sc_buggy_t["init"], label="init")
ax1.plot(sc_buggy_t["d"], sc_buggy_t["config"], label="config")
ax1.plot(sc_buggy_t["d"], sc_buggy_t["cons_gen"], label="cons_gen")

ax1.set_title("Buugy Decoder: Execution Time visualization", fontsize=18)
ax1.set_xlabel("Code distance (d)", fontsize=14)
ax1.set_ylabel("Execution Time(in sec)", fontsize=14, color=color1)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis


ax2.plot(sc_buggy_t["d"], sc_buggy_t["cons_sol"], label="cons_sol")

ax1.set_ylabel("Execution Time(in sec)", fontsize=14, color=color1)




# plt.legend()
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width, pos.height * 0.85])
ax.legend(
    loc='upper center', 
    bbox_to_anchor=(0.5, 1.35),
    ncol=4, 
)

plt.show()

plt.rcParams.update(plt.rcParamsDefault)
```

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

fig, ax = plt.subplots()

# plt.rcParams.update({'font.size': 20})
ax.plot(sc_buggy_t["d"], sc_buggy_t["init"], label="init")
ax.plot(sc_buggy_t["d"], sc_buggy_t["config"], label="config")
ax.plot(sc_buggy_t["d"], sc_buggy_t["cons_gen"], label="cons_gen")

# plt.plot(sc_buggy_t["d"], sc_buggy_t["all"], label="all")

# plt.scatter(sc_correct["d"], sc_correct["all"], label="correct")
# plt.scatter(sc_buggy["d"], sc_buggy["all"], label="buggy")

# plt.scatter(sc_buggy["d"], sc_buggy["all"])

ax.set_title("Buggy Decoder: Visualize Non Dominating Stages")
ax.set_xlabel("Code distance (d)")
ax.set_ylabel("Execution Time(in sec)")

# plt.legend()
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width, pos.height * 0.85])
ax.legend(
    loc='upper center', 
    bbox_to_anchor=(0.5, 1.35),
    ncol=4, 
)
```
